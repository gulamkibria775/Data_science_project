```python
import pandas as pd
data=pd.read_csv("/content/driver/MyDrive/All_data_file/sonar/sonar.csv",header=None)
data.head()
```

```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
tf.compat.v1.disable_eager_execution()
```

```python
print(len(data.columns))
```

```python
x=data[data.columns[0:60]].values
y=data[data.columns[60]]
```

```python
y_label=LabelEncoder()
y=y_label.fit_transform(y)
print(y)
```

```python
y=y.reshape(-1,1)
y
```

```python
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('my_ohe', OneHotEncoder(), [0])], remainder='passthrough')
Y = ct.fit_transform(y) #.toarray()
print(Y)
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,Y,test_size=0.20, random_state=42)
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
learning_rate=.01
traning_epochs=100
cost_history=np.empty(shape=[1],dtype=float)
```

```python
n_class=2
x=tf.compat.v1.placeholder(tf.float32,[None,60])
w=tf.Variable(tf.zeros([60,n_class]))
b=tf.Variable(tf.zeros([n_class]))
```

```python
print(x)
print(w)
print(b)
```

```python
init=tf.compat.v1.global_variables_initializer()
y=tf.compat.v1.placeholder(tf.float32,[None,n_class])
pred=tf.nn.softmax(tf.matmul(x,w)+b)
```

```python
cost_function=tf.reduce_mean(-tf.reduce_sum((y*tf.math.log(pred))))
```

```python
train_step=tf.compat.v1.train.GradientDescentOptimizer(learning_rate).minimize(cost_function)
```

```python
sess=tf.compat.v1.Session()
sess.run(init)
mse_history=[]
```

```python
for i in range(100):
    sess.run(train_step,feed_dict=({x: train_x, y: train_y}))
    cost = sess.run(cost_function,feed_dict={x: train_x,y:train_y})
    cost_history = np.append(cost_history,cost) #cost history we have created previously and using it here just to store the cost of each epoch

    #after above 3 lines weights get updated and once we have updated weights we will evaluate the test data set.

    pred_y = sess.run(pred, feed_dict={x: test_x})  # y is a linear model which is using softmax function. Once we update the weights we will evaluate it on the test_x
    print('epoch : ', i,  ' - ', 'cost: ', cost)  # printing cost or error associated with each epoch

    # we want to calculate the mean square error now
    mse = tf.reduce_mean(tf.square(pred_y - test_y))  # test_y is the actual output and pred_y is the predicted output
    mse_history.append(sess.run(mse))
```

```python
correct_prediction=tf.equal(tf.argmax(pred,1),tf.argmax(y,1))
accuracy=tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
print("Accuracy",(sess.run(accuracy,feed_dict={x:test_x,y:test_y})))
```
